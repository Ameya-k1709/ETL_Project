```python
import os, json
import pandas as pd
import numpy as np
import glob
```

```python
#Take all the json files as input and store them into a device files list
#Created a function get_files() to create a json files list.

def get_files(filepath):
    all_files = []
    for root, dirs, files in os.walk(filepath):
        files = glob.glob(os.path.join(root,"*json"))
        for f in files:
            all_files.append(os.path.abspath(f))
    return all_files
```
